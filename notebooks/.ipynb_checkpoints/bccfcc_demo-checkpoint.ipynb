# A Bayesian Framework for Persistent Homology

We provide a tutorial for our Python module that implements a point process model for Bayesian inference with persistence diagrams. Please see our original paper for details: https://epubs.siam.org/doi/pdf/10.1137/19M1268719.

We'll consider a classification problem from atom probe tomography (APT). Our goal is to build a model that distinguishes between body-centered cubic (BCC) and face-centered cubic (FCC) atomic neighborhoods. Because these two classes have notably distinct geometric and topological properties, persistence diagrams are a useful tool for this task and we choose to work in persistence diagram space.

![SegmentLocal](bccfcc_figs/rips.gif "segment")

<img src="bccfcc_figs/bccfcc_pd_demo.png">

In practice, APT measurements can be sparse and noisy. Some experiments have about 67% of data missing, and this data is further corrupted by experimental noise. This means if we want to work in persistence diagram space, we need a probabilistic model to account for uncertainty in the original data. 

<img src="bccfcc_figs/bcc_v_fcc.png">

We use a Bayesian framework for inference. At a high level, the Bayesian pipeline consists of prior assumptions to regularize the final inference model (known in the Bayesian framework as the posterior) and a likelihood that tells us how to update our prior assumptions based on data. 

# Bayesian Workflow

We illustrate the statiscal underpinnings of our method with a standard TDA example, the noisy circle.

```python
from bayes_tda.demos import Circle

noisy_circle = Circle()
noisy_circle.show()
```

Our Bayesian model views a persistence diagram $\mathcal{D}$ as a sample from a point process $D$, which is a random collection of elements whose cardinality may vary. A useful summary statistic of a point process is its intensity,$\lambda$, which satisfies $\mathbb{E}(|D \cap \mathcal{A}|) = \int _{A}\lambda(x)\,dx$ for every $A \subset \mathbb{W} := \{(b,p) \in \mathbb{R}^2: b, p > 0\}$. Our Bayesian model gives a recipe to compute posterior intensities using prior intensities, a likelihood, and observed persistence diagrams. In our implementation, the prior intensity $\lambda _D(x)$ is given by a Gaussian mixture.

\begin{equation}
\lambda_{D}(x) = \sum_{i=1}^{N} w_i\mathcal{N}^{'}(x;\mu_i,\sigma_i I)
\end{equation}

where $N$ is the number of mixture components, $w_i$ are weights, and 
\begin{equation}
\mathcal{N}^{'}(x;\mu_i,\sigma_i I) = \begin{cases}\mathcal{N}(x;\mu_i,\sigma_i I), x \in \mathbb{W} \newline
                                                    0, \mathrm{otherwise}
                                    \end{cases}
\end{equation}
where $\mathcal{N}(x;\mu_i,\sigma_i I)$ is the Gaussian density, with mean $\mu _i$ and covariance matrix $\sigma _i I$. Thus, we can create prior intensities by specifying the means, covariances, and weights of each mixture component, respectively. Below, we demonstrate the syntax for doing so.

```python
from bayes_tda.intensities import RGaussianMixture
import numpy as np

# parameters of the prior
prior_mus = np.array([[1, 3],
                    [2, 0.5]]) # means of prior
                  
prior_sigmas = np.array([10,
                  50]) # variances in each component of prior
                  
prior_weights = np.array([1,
                    1]) # weights of each component in prior

prior = RGaussianMixture(prior_mus, prior_sigmas, prior_weights)

# visualize prior
linear_grid = np.linspace(0, 4, 28)
prior.show_density(linear_grid, title = 'Prior Intensity', cmap = 'coolwarm')
```

Another component of our model is the clutter, i.e. an intensity for unanticipated features that measures the probability of features arising from noise instead of characteristic topology in the data. In the circle example, these features are the low persistence features. 

The clutter is another Gaussian mixture supplied by the user and it serves as another source of prior information that prevents overfitting the posteriors to noise. Since we anticipate that features whose persistences are close to zero are the result of noise, we leverage this information to define the clutter.  

```python
clutter_mus = np.array([[0.15,0.1],
                        [1,0.3],
                        [0.5,0.5]]) # unanticipated features means

clutter_sigmas = np.array([0.1,0.1,0.1]) # unanticipated features covariance magnitudes

clutter_weights = np.array([2,
                            2,
                            2]) # unanticipated features weights

clutter = RGaussianMixture(clutter_mus, clutter_sigmas, clutter_weights)

# visualize clutter
linear_grid = np.linspace(0, 4, 28)
clutter.show_density(linear_grid, title = 'Clutter Intensity', cmap = 'coolwarm')
```

The last ingredient we need to compute a posterior distribution is a likelihood $\ell$, which in our implementation is given by a restricted Gaussian density, 
\begin{equation}
\ell(y|x) = \mathcal{N}^{'}(y;x,\sigma_{D_{Y_{O}}} I)
\end{equation}

The bandwidth parameter $\sigma_{D_{Y_{O}}}$ controls the level of contribution that the observed data has to the posterior. Lower values lead to posterior densities that are more highly concentrated around observed data points. 

```python
from bayes_tda.intensities import Posterior

DYO = [noisy_circle.dgm] # list of observed persistence diagrams
sigma_DYO = 0.3 # magnitude of covariance matrix for likelihood
posterior = Posterior(DYO, prior, clutter, sigma_DYO)

posterior.show_prior(linear_grid, cmap = 'coolwarm')
posterior.show_clutter(linear_grid, cmap = 'coolwarm')
posterior.show_lambda_DYO(linear_grid, show_means = False, cmap = 'coolwarm',
                         plot_additional_pts = True,
                         additional_pts = noisy_circle.dgm)
```

# Machine Learning Example: BCC-FCC Classification

Finally, we revisit our machine learning example. Recall our problem: distinguish body-centered cubic (BCC) from face-centered cubic (FCC) atomic neighborhoods. We consider 1000 atomic point clouds each, about 500 BCC and FCC, respectively, with 67% sparsity and corrupted by measurement noise.


BCC
<img src="bccfcc_figs/bcc.png">

FCC
<img src="bccfcc_figs/fcc.png">

Given a persistence diagram $\mathcal{D}$, we estimate the Bayes factor for the competing hypotheses: (i) $\mathcal{D}$ corresponds to a BCC sample and (ii) $\mathcal{D}$ corresponds to a FCC sample. In Bayesian inference, this can be qunatified with a Bayes factor: 
$$BF(\mathcal{D}) := \frac{\mathbb{P}(\mathcal{D}|\mathrm{BCC})}{\mathbb{P}(\mathcal{D}|\mathrm{FCC})}.$$ 

Large and small values of $BF(\mathcal{D})$ provide strong evidence for (i) and (ii), respectively. The Bayes factor can be used for classification, as summarized in the following schematic.
<img src="bccfcc_figs/bcc_v_fcc.png">

We implement as Bayes factor classification algorithm in a Python class. The inputs to our algorithm are the hyperparameters in the prior and clutter, the likelihood bandwidth $\sigma_{D_{Y_{O}}}$, and the proportion of training data to used to compute posteriors for each class. 

```python
from bayes_tda.classifiers import EmpBayesFactorClassifier as EBFC


# load data
DATA_PATH = '/home/chris/projects/bayes_tda/data/'
DATA = 'bccfcc_small.npy'
LABELS = 'bccfcc_small_labels.npy'

data = np.load(DATA_PATH + DATA, allow_pickle = True)
labels = np.load(DATA_PATH + LABELS, allow_pickle = True)

# Set inputs to algorithm
PRIOR_MUS = np.array([[2, 2]])

PRIOR_SIGMAS = np.array([20])

PRIOR_WEIGHTS = np.array([1])

CLUTTER_MUS = np.array([[0, 0],
                      [0.1, 0.02]])

CLUTTER_SIGMAS = np.array([0.1,
                         0.1])

CLUTTER_WEIGHTS = np.array([0,
                          0])

SIGMA_DYO = 0.1

PRIOR_PROP = 0.2


    
# build prior and clutter
prior = RGaussianMixture( mus = PRIOR_MUS, 
                             sigmas = PRIOR_SIGMAS, 
                             weights = PRIOR_WEIGHTS, 
                             normalize_weights= False)
    
clutter = RGaussianMixture(mus = PRIOR_MUS, 
                             sigmas = PRIOR_SIGMAS, 
                             weights = PRIOR_WEIGHTS, 
                             normalize_weights= False)
    
```

Now, we can perform our classification algortihm with a few lines of code. 

```python
# build classifier
from bayes_tda.classifiers import EmpBayesFactorClassifier as EBFC

classifier = EBFC(data = data,
                      labels = labels,
                      data_type = 'diagrams')
    
scores = classifier.compute_scores(clutter, 
                                       prior,
                                       prior_prop = PRIOR_PROP,
                                       sigma_DYO = SIGMA_DYO)
    
```

Finally, we examine our results. We show the Bayes factors for each class and report the Area Under the receiver operating characteric Curve (AUC) of our classifier. 

```python
# examine score distributions
from bayes_tda.bccfcc_post import _show_bayes_factor_distribution

auc = _show_bayes_factor_distribution(scores)
print('AUC: ' + str(auc))
```

We observe higher Bayes factors for the BCC class as expected and an AUC around 0.87. Not bad for such noisy data.
